```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pid
data_df=pd.read_csv("testset.csv")
data_df.columns
```

```python
data_df['datetime_utc'] = pd.to_datetime(data_df.datetime_utc) 
data_df['month']=data_df['datetime_utc'].dt.month
```

```python
new_names = {" _conds": "_conds", 
                    " _dewptm":"_dewptm",
                    " _fog":"_fog",
                    " _hail":"_hail",
                    " _heatindexm":"_heatindexm",
                    " _hum":"_hum",
                    " _precipm":"_precipm",
                    " _pressurem":"_pressurem",
                    " _rain":"_rain",
                    " _snow":"_snow",
                    " _tempm":"_tempm",
                    " _thunder":"_thunder",
                    " _tornado":"_tornado",
                    " _vism":"_vism",
                    " _wdird":"_wdird",
                    " _wdire":"_wdire",
                    " _wgustm":"_wgustm",
                    " _windchillm":"_windchilm",
                    " _wspdm":"_wspdm"}
```

```python
mean_temp=round(data_df["_tempm"].mean(),2)
data_df["_tempm"].fillna(mean_temp,inplace=True)
```

```python
mean_vism=round(data_df["_vism"].mean(),2)
data_df["_vism"].fillna(mean_vism,inplace=True)
```

```python
cond = data_df['_conds']
```

```python
freq = {} # Nov 1997
for ind,condition in cond.items():
    mo = str(ind.month)
    ye = str(ind.year)
    if ye == '1997' and mo == '11':
        if freq.get(condition) == None:
            freq[condition] = 0
        freq[condition] += 1

y = list(freq.values())
x = list(freq.keys())
pid.pie(y,labels=x)
pid.show()
```

```python
freq = {} # April 2017

for ind,condition in cond.items():
    mo = str(ind.month)
    ye = str(ind.year)
    if ye == '2017' and mo == '4':
        if freq.get(condition) == None:
            freq[condition] = 0
        freq[condition] += 1

y = list(freq.values())
x = list(freq.keys())
pid.pie(y,labels=x)
pid.show()
```
